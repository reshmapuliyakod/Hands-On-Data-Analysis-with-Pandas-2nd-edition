```python
import pandas as pd
```

```python
weather = pd.read_csv('data/nyc_weather_2018.csv')
weather.head()
```

Querying DataFrames

```python
# Below code is very similar to the Select * from table where in SQL
snow_data = weather.query('datatype=="SNOW" and value>0 and station.str.contains("US1NY")')
snow_data.head()
```

```python
station_info = pd.read_csv('data/weather_stations.csv')
station_info.head()
```

```python
station_info.id.describe()
```

```python
weather.station.describe()
```

```python
def get_row_count(*dfs):
    return [df.shape[0] for df in dfs]

get_row_count(station_info,weather)
```

Joining or Merging Dataframes ( similar to join operations in sql) - merge()
by default , merge() - inner join by default

```python
inner_join = weather.merge(
    station_info, left_on= 'station', right_on='id'
)
```

```python
inner_join.sample(5,random_state=0)
```

```python
# both station and id columns are returned above even they are same
weather.merge(
station_info.rename(dict(id='station'),axis=1),
on ='station'
).sample(5,random_state=0)
```

```python
left_join = station_info.merge(

weather, left_on='id', right_on='station', how='left'

)

left_join.head()
```

```python
right_join = weather.merge(
station_info, left_on='station',right_on='id',how='right'
)
right_join.head()
```

```python
get_row_count(inner_join,left_join,right_join)
```

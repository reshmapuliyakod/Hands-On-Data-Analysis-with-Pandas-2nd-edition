```python
import numpy as np
import pandas as pd
```

```python
faang = pd.DataFrame()
for ticker in ['aapl','amzn','fb','goog','nflx']:
    df = pd.read_csv(f'exercises\\{ticker}.csv')
    df.insert(0,'ticker',ticker.upper())
    faang = faang.append(df)
faang.to_csv('exercises\\faang.csv',index =False)
faang.head()
```

```python
faang = faang.assign(
    date = pd.to_datetime(faang.date),
    volume = faang.volume.astype('int')

).sort_values(['date','ticker'])

faang.head()
```

```python
faang.nsmallest(7, 'volume')
```

```python
faang.head()
```

```python
faang_melted = faang.melt(id_vars=('date','ticker'),value_vars=['high','low','open','close','volume'],
value_name='values',
var_name='facts')
faang_melted.head()
```
